# Results over re-runs of `word2vec`
Multiple runs over different samples of the same size (15% and 50% of wikipedia)

First 3 samples are random (the first isn't really- it's consecutive files). The last item is averages over the previous 3

```python
%cd ~/NetBeansProjects/ExpLosion/
from notebooks.common_imports import *
from thesisgenerator.utils.output_utils import *

sns.set_context("paper")
```

```python
d = {'vectors__rep__in': [0, 1, 2, -1], # todo add -1 ?????
     'document_features__in': ['AN_NN'], 
     'vectors__unlabelled_percentage__in': [15, 50], 
     'labelled__in': ['amazon_grouped-tagged'], 
     'vectors__unlabelled__in': ['wiki'], 
     'vectors__algorithm__in': ['word2vec'],
     'decode_handler__in': ['SignifiedOnlyFeatureHandler'], 
     'k__in': ['3'], 
     'noise__in': ['0.0']}
exp_ids = Experiment.objects.values_list('id', flat=True).filter(**d)
scores, folds, ids = get_scores(exp_ids)
assert ids # we succeeded in getting results

df = pd.DataFrame({'F1': scores,
                   'cvfold': folds,
                   'rep': get_vectors_field(ids, 'rep'),
                   'percent': get_vectors_field(ids, 'unlabelled_percentage'),
                   'composer': get_vectors_field(ids, 'composer')})
df[(df.cvfold==0) & (df.composer=='Add')].to_html('tmp.html')
df[(df.percent==50) & (df.composer=='Add')].shape
# df = df[(df.cvfold==0)]
# df.to_html('tmp.html')
```

```python
g = sns.FacetGrid(df,  col='percent', sharey=True);
g.map(sns.boxplot, x='composer', y='F1', data=df);
g.map(sns.stripplot, x='composer', y='F1', 
      hue='rep', hue_order=[0, 1, 2, -1], data=df, 
      size=5, jitter=0.2, palette='muted', edgecolor='grey', alpha=.8);
```
